```python
import pandas as pd
from itranslate import itranslate 
```

```python
Lang_Trans = pd.read_csv("English.csv")
Lang_Trans.head()
```

```python
Lang_Trans.shape
```

```python
Lang_Trans.duplicated().sum()
```

```python
Lang_Trans = Lang_Trans.drop_duplicates(keep='first',ignore_index=True)
Lang_Trans.shape
```

# 1. Here I taken only 500 samples because of runtime error,there are huge data in the provided csv , i tried with textblob,google_trans,itrans etc... I am facing runtime error for all the above tools hence i translated only 500 data to french,spanish and german

```python
Translations = Lang_Trans[:500]
Translations["Spanish"] = Translations["English words/sentences"].apply(lambda x : itranslate(x, to_lang = 'es'))
Translations
```

```python
Translations["French"] = Translations["English words/sentences"].apply(lambda x : itranslate(x, to_lang = 'fr'))
Translations
```

```python
Translations["German"] = Translations["English words/sentences"].apply(lambda x : itranslate(x, to_lang = 'de'))
Translations
```
